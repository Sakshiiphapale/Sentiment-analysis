```python
#sakshi phapale
from googleapiclient.discovery import build
import pandas as pd

# YouTube Data API key
api_key = "AIzaSyAF1eLG7qZkk4nF3p-ffzxXdRKa79p4fo0"

# YouTube video ID
video_id = "0mAz5QhCLYo"

# Function to fetch YouTube comments
def get_youtube_comments(api_key, video_id):
    youtube = build('youtube', 'v3', developerKey=api_key)
    response = youtube.commentThreads().list(
        part='snippet',
        videoId=video_id,
        maxResults=100
    ).execute()

    comments = []
    for item in response['items']:
        comment = item['snippet']['topLevelComment']['snippet']
        comments.append({
            'author': comment['authorDisplayName'],
            'comment': comment['textDisplay']
        })

    return comments

# Fetch YouTube comments
comments = get_youtube_comments(api_key, video_id)

# Convert comments to DataFrame
comments_df = pd.DataFrame(comments)
```

```python
#sakshi phaaple
from textblob import TextBlob

# Function for sentiment analysis
def analyze_sentiment(comment):
    analysis = TextBlob(comment)
    sentiment = analysis.sentiment.polarity
    return sentiment

# Apply sentiment analysis to comments
comments_df['sentiment'] = comments_df['comment'].apply(analyze_sentiment)

```
